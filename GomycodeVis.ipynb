```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("used cars.csv")
```

```python
#Create a bar chart using Plotly :
fig = px.bar(df, x="Brand", y="Cars Listings", color="Cars Listings", height=400)
```

```python
#Add a title and labels to the x and y axis
fig.update_layout(title_text="Cars Listings by Brand", xaxis_title="Brand", yaxis_title="Cars Listings")
```

```python
#Customize the color scale and add a color bar :
fig.update_traces(marker=dict(color='red', opacity=0.7, line=dict(color='red', width=1.5)))
fig.update_layout(coloraxis_colorbar=dict(title="Cars Listings"))
```

```python
#Show the bar chart
fig.show()
```

```python
#Create a pie chart using Plotly Express, specifying the data frame, the column for the values, the column for the names, and the color for each slice.

fig = px.pie(df, values='Cars Listings', names='Brand', color='Brand', title='Engine Fuel Type Distribution of Cars')

#Update the text position of the labels to be inside each slice and show the percent and label information

fig.update_traces(textposition='inside', textinfo='percent+label')

#Update the layout to set a minimum size for the text and to hide any overlapping text

fig.update_layout(uniformtext_minsize=14) #uniformtext_mode='hide')

#Show the final chart

fig.show()
```

```python
scatter = pd.read_csv("scatter_data.csv")
scatter
```

```python
# how to create scatter plot using plotly :
# In this section, we will learn how to create a scatterplot using real estate data, specifically using the price and area features of each property.

# Import plotly.graph_objs and plotly.offline modules. plotly.graph_objs module provides different types of charts, while plotly.offline is used to create interactive charts offline.

import plotly.graph_objs as go
import plotly.offline as pyo

# Create a trace using the go.Scatter() function. This will create a scatter plot. Here, x and y arguments are the data to be plotted on the X and Y axis, respectively. The mode argument specifies the type of plot, in this case, markers. The marker argument specifies the color variable, which is based on the Building Type column of the DataFrame.

trace = go.Scatter(x=scatter['Area (ft.)'], y=scatter['Price'], mode='markers', marker=dict(color=scatter['Building Type']))

#Create a data list. Here, the trace object created in the previous step is added to a list called data.

data = [trace]

# Create a layout object using the go.Layout() function. This will create a layout for the chart, including the title, X and Y axis labels.

layout = go.Layout(title='Year of Sale vs. Price',xaxis=dict(title='Area (ft.)'),yaxis=dict(title='Price'))

# Create a figure using the go.Figure() function. This combines the data and layout objects to create a complete figure.

fig = go.Figure(data=data, layout=layout)

# Show the plot using pyo.plot() function. This function takes the fig object as an argument and displays the plot in the default web browser.

pyo.plot(fig)
```
